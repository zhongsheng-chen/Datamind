# 测试/predict_label

```python
import requests

url = "http://127.0.0.1:3000/predict_label"
headers = {"Content-Type": "application/json"}

payload = {
    "age": 35,
    "income": 12000,
    "debt_ratio": 0.3,
    "loan_amount": 5000,
    "existing_loans": 2,
    "total_tax_records": 5,
    "total_tax_amount": 20000,
    "avg_tax_amount": 4000,
    "max_tax_amount": 6000,
    "min_tax_amount": 2000,
    "tax_amount_std": 1500,
    "loan_to_income_ratio": 0.42,
    "existing_loans_ratio": 0.1
}

response = requests.post(url, json=payload, headers=headers)

print("Status code:", response.status_code)
print("Response:", response.json())
```
